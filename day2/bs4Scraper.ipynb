```python
# Import the requests library to send HTTP requests
import requests

# Import BeautifulSoup to parse and navigate the HTML content
from bs4 import BeautifulSoup

# Define the URL of the fake jobs listing page (demo site by Real Python)
url = "https://realpython.github.io/fake-jobs/"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the response HTML content using BeautifulSoup with the built-in HTML parser
soup = BeautifulSoup(response.content, "html.parser")

# Find all <div> elements with class "card-content"
# Each "card-content" div contains information about one job posting
jobs = soup.find_all("div", class_="card-content")
```

# **List All Unique Companies**

```python
# Use a set comprehension to extract company names from all job entries in 'jobs'
# job.find("h3", class_="company") finds the <h3> tag with class 'company'
# .text.strip() gets the visible text and removes extra whitespace
# Using a set {} ensures that each company name is stored only once (no duplicates)
companies = {
    job.find("h3", class_="company").text.strip()
    for job in jobs
}
print(companies)
# Loop through the set of unique company names
# for company in companies:
#     # Print each company name
#     print(company)
```

# **Extract All Job Titles**

```python
# Loop through each job element in the list called 'jobs'
# Each item in 'jobs' is expected to be a BeautifulSoup object representing a job posting
for job in jobs:

    # Find the <h2> tag with class 'title' inside the current job element
    # Extract the text content from the tag using .text
    # Use .strip() to remove any leading or trailing whitespace (like newlines or spaces)
    title = job.find("h2", class_="title").text.strip()

    # Print the cleaned job title to the console
    print(title)
```

# **Extract Company and Location**

```python
# Loop through all <div> elements with class "card"
# Each "card" represents an individual job posting on the page
for job in soup.find_all("div", class_="card"):

    # Look for a <time> element inside the current job card
    # This usually contains the posting date (e.g., <time datetime="2024-01-01">)
    date_span = job.find("time")

    # Find the <h2> element with class "title" and extract the text content
    # .text gets the text inside the tag, and .strip() removes extra whitespace
    title = job.find("h2", class_="title").text.strip()

    # Check if a <time> tag was found in this job card
    if date_span:
        # Print the job title and the posting date using the 'datetime' attribute of <time>
        print(f"{title} - Posted on: {date_span['datetime']}")
```

# Extract Job Posting Date

```python
# Loop through all <div> elements with class "card"
# Each card represents an individual job posting
for job in soup.find_all("div", class_="card"):

    # Look for the <time> element inside the job card
    # This usually contains the posting date
    date_span = job.find("time")

    # Find the <h2> tag with class "title" and extract its text content
    # .text gets the inner text, and .strip() removes leading/trailing whitespace
    title = job.find("h2", class_="title").text.strip()

    # If a <time> element was found (i.e., the posting date exists)
    if date_span:
        # Access the 'datetime' attribute of the <time> tag, which contains the post date
        # Print the job title along with the posting date
        print(f"{title} - Posted on: {date_span['datetime']}")
```

# **Extract Job Apply Links**

```python
# Loop through all <footer> tags with class 'card-footer'
# These are assumed to contain the "Apply" button/link for each job card
for job in soup.find_all("footer", class_="card-footer"):

    # Find the <a> tag with exact text "Apply" inside the current footer
    # .get("href") extracts the URL value from the href attribute
    apply_link = job.find("a", text="Apply").get("href")

    # Print the extracted apply link to the console
    print(apply_link)
```

# Save All Job Info to CSV

```python
# Import the built-in csv module to work with CSV files
import csv

# Open a new file named 'jobs.csv' in write mode with UTF-8 encoding
# 'newline=""' ensures no extra blank lines are added on Windows
with open("jobs.csv", "w", newline="", encoding="utf-8") as f:

    # Create a CSV writer object that will write to the file
    writer = csv.writer(f)

    # Write the header row to the CSV file
    writer.writerow(["Title", "Company", "Location"])

    # Loop through each job element in the jobs list
    for job in jobs:

        # Extract the job title from the <h2> tag with class 'title' and strip whitespace
        title = job.find("h2", class_="title").text.strip()

        # Extract the company name from the <h3> tag with class 'company' and strip whitespace
        company = job.find("h3", class_="company").text.strip()

        # Extract the job location from the <p> tag with class 'location' and strip whitespace
        location = job.find("p", class_="location").text.strip()

        # Write the extracted job data as a new row in the CSV file
        writer.writerow([title, company, location])
```

# **Filter Jobs by Keyword ("Python")**

```python
# Loop through each job element in the jobs list
for job in jobs:

    # Extract the job title from the <h2> tag with class "title", remove leading/trailing spaces
    title = job.find("h2", class_="title").text.strip()

    # Check if the word "python" appears in the job title (case-insensitive)
    if "python" in title.lower():

        # Extract the company name from the <h3> tag with class "company", remove spaces
        company = job.find("h3", class_="company").text.strip()

        # Extract the job location from the <p> tag with class "location", remove spaces
        location = job.find("p", class_="location").text.strip()

        # Print the job title, company, and location in formatted output
        print(f"{title} - {company} - {location}")
```

# **Latest Tech News Headlines**

```python
# Import the requests library to make HTTP requests
import requests

# Import BeautifulSoup for parsing HTML content
from bs4 import BeautifulSoup

# Import pandas to store and manipulate data in tabular form
import pandas as pd

# Import datetime to timestamp when data was scraped
import datetime

# Define the target URL for scraping news headlines
url = "https://www.hindustantimes.com/india-news"

# Set up headers to mimic a real browser (helps prevent blocking)
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Send a GET request to the URL with headers
response = requests.get(url, headers=headers)

# Parse the HTML content using BeautifulSoup with the default HTML parser
soup = BeautifulSoup(response.content, "html.parser")

# Initialize an empty list to store the extracted headlines
headlines = []

# Loop through all <a> tags inside <h3> tags (which usually contain news titles)
for item in soup.select("h3 a"):
    # Extract the visible title text and strip any surrounding whitespace
    title = item.get_text(strip=True)

    # Extract the link from the 'href' attribute
    link = item.get("href")

    # If the link is relative (starts without http), prepend the full domain
    if not link.startswith("http"):
        link = "https://www.hindustantimes.com" + link

    # Append a dictionary of title and full URL to the headlines list
    headlines.append({"Title": title, "URL": link})

# Convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(headlines)

# Add a new column with the current timestamp to record when the data was scraped
df["Scraped_At"] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Define the filename for saving the scraped data as a CSV
csv_filename = "ht_india_news.csv"

# Save the DataFrame to a CSV file without the index column, using UTF-8 encoding
df.to_csv(csv_filename, index=False, encoding="utf-8")

# Print the number of headlines scraped and the filename
print(f"{len(df)} headlines scraped and saved to {csv_filename}")
```
